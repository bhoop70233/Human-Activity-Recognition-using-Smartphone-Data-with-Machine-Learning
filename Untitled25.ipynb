```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv("train-1.csv")
test = pd.read_csv("test.csv")
```

```python
train['Data']='Train'
test['Data']='Test'
both=pd.concat([train,test],axis=0).reset_index(drop=True)
both['subject']='#' +both['subject'].astype(str)
```

```python
def basic_details(df):
  b=pd.DataFrame()
  b['Missing value']=df.isnull().sum()
  b['N unique']=df.nunique()
  b['dtype']=df.dtypes
  return b
basic_details(both)
```

```python
activity=both['Activity']
label_counts=activity.value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=label_counts.index,y=label_counts )
```

```python
Data = both['Data']
Subject = both['subject']
train = both.copy()
train = train.drop(['Data','subject','Activity'], axis =1)
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
slc = StandardScaler()
train = slc.fit_transform(train)
```

```python
# dimensionality reduction
import pandas as pd # Import pandas if not already imported
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Handle missing values (replace with mean for example)
imputer = SimpleImputer(strategy='mean')
train = imputer.fit_transform(train)

pca = PCA(n_components=10)
train = pca.fit_transform(train)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, activity, test_size = 0.2, random_state = 0)
```

```python
num_folds = 10
seed = 0
scoring = 'accuracy'
results = {}
accuracy = {}
```

```python
# Finalizing the model and comparing the test, predict results
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Handle NaN values in y_train and y_test
y_train = y_train.fillna('Unknown') # Replace NaN with a placeholder
y_test = y_test.fillna('Unknown')

# Fit and transform y_train to numerical labels
y_train_encoded = le.fit_transform(y_train)

# Transform y_test using the same encoder
y_test_encoded = le.transform(y_test)

model = KNeighborsClassifier(algorithm= 'auto', n_neighbors= 8, p= 1, weights= 'distance')

_ = cross_val_score(model, X_train, y_train_encoded, cv=10, scoring=scoring) # Use encoded y_train
results["GScv"] = (_.mean(), _.std())

model.fit(X_train, y_train_encoded) # Use encoded y_train
y_predict = model.predict(X_test)

# Calculate accuracy using encoded y_test
accuracy["GScv"] = accuracy_score(y_test_encoded, y_predict)

print(classification_report(y_test_encoded, y_predict)) # Use encoded y_test

cm= confusion_matrix(y_test_encoded, y_predict) # Use encoded y_test
sns.heatmap(cm, annot=True)
```
